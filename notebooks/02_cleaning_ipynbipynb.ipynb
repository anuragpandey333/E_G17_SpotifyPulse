```python
import pandas as pd
df=pd.read_csv("/content/spotify_user_behavior_realistic_50000_rows.csv")
```

```python
df.head()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
```

```python
df['country'] = df['country'].str.lower().str.strip()
```

```python
df['subscription_type'] = df['subscription_type'].str.lower().str.strip()
```

```python
df.isnull().sum()
```

```python
df['age'].mean()
df['age'].median()
```

```python
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

```python
df = df[(df['age']>=lower_bound) & (df['age']<=upper_bound)]
```

```python
df['age']=df['age'].fillna(df['age'].mean())
df['country']=df['country'].fillna('unknown')
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['age_group'] = 'unknown'
df.loc[df['age'] < 12, 'age_group'] = 'child'
df.loc[(df['age'] >= 12) & (df['age'] < 20), 'age_group'] = 'teenager'
df.loc[(df['age'] >= 20) & (df['age'] < 25), 'age_group'] = 'young_adult'
df.loc[(df['age'] >= 25) & (df['age'] < 35), 'age_group'] = 'adult'
df.loc[(df['age'] >= 35) & (df['age'] < 50), 'age_group'] = 'mid_age'
df.loc[df['age'] >= 50, 'age_group'] = 'senior'
```

```python
df['inactive_3_months_flag'] = (df['months_inactive'] >= 3).astype(int)
```

```python
# df['ad_interaction'] = df['ad_interaction'].map({'Yes': 1, 'No': 0})
# df['ad_conversion_to_subscription'] = df['ad_conversion_to_subscription'].map({'Yes': 1, 'No': 0})
```

```python
df['engagement_score'] = (
    df['avg_listening_hours_per_week'] * 0.4 +
    df['playlists_created'] * 0.3 -
    df['avg_skips_per_day'] * 0.3
).round(3)
df['engagement_level'] = pd.qcut(df['engagement_score'], 3, labels=['Low', 'Medium', 'High'])
```

```python
df['subscription_type'] = df['subscription_type'].str.lower().str.strip()
df['most_liked_feature'] = df['most_liked_feature'].str.lower().str.strip()
```

```python
df.to_csv("spotify_cleaned.csv", index=False)
```
