```python
import pandas as pd
df = pd.read_csv('spotify_cleaned.csv')
df.head()
df.info()
```

```python
import pandas as pd

total_users = len(df)

kpis = {}

    # ---------------- RETENTION ---------------- #
kpis['churn_rate'] = df['inactive_3_months_flag'].mean()
kpis['churn_rate']

```

```python
kpis['active_users_pct'] = 1 - kpis['churn_rate']
kpis['active_users_pct']
```

```python
kpis['at_risk_users_pct'] = (
    (
        (df['inactive_3_months_flag'] != 1) &  # not churned
        (df['months_inactive'] >= 2)           # at risk condition
    ).mean() * 100
)
kpis['at_risk_users_pct']
```

```python
kpis['avg_inactivity_months'] = df['months_inactive'].mean()
kpis['avg_inactivity_months']
```

```python
kpis['avg_listening_hours'] = df['avg_listening_hours_per_week'].mean()
kpis['avg_listening_hours']
```

```python
kpis['avg_engagement_score'] = df['engagement_score'].mean()
kpis['avg_engagement_score']
```

```python
kpis['avg_skips_per_day'] = df['avg_skips_per_day'].mean()
kpis['avg_skips_per_day']
```

```python
kpis['total_playlists_created'] = df['playlists_created'].sum()
kpis['total_playlists_created']
```

```python
kpis['premium_conversion_rate'] = (
        (df['subscription_type'].str.lower()!='free').mean()
    )
kpis['premium_conversion_rate']
```

```python
kpis['free_user_ratio'] = (
        (df['subscription_type'].str.lower() == 'free').mean()
    )
kpis['free_user_ratio']
```

```python
kpis['avg_playlists_per_user'] = df['playlists_created'].mean()
kpis['avg_playlists_per_user']
```

```python
if 'music_suggestion_rating_1_to_5' in df.columns:
    kpis['feature_interaction_score'] = df['music_suggestion_rating_1_to_5'].mean()

else:
    kpis['feature_interaction_score'] = None
kpis['feature_interaction_score']
```
